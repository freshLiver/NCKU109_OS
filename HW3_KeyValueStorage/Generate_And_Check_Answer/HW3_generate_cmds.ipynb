```
from random import randint, choice, sample

ALLOW_CHARS = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"
INT64MAX = 9223372036854775807

# key   : [0:2**63-1]
# value : always 128 digits of [A-Za-z0-9]

# PUT   : PUT key value
# GET   : GET key
# SCAN  : SCAN key1 key2

available_keys = sample( range(0,INT64MAX), 100000 )
```

```
def randKey( init = 0 ) -> int:
    return choice(available_keys)

def rand2Key(limit) -> (int,int):
    key1 = randKey()
    key2 = min( key1+randint(1,limit), INT64MAX )
    return (key1,key2)

def randValue() -> str:
    cmd = ""
    for i in range(128):
        cmd += choice(ALLOW_CHARS)
    return cmd
```

```
cmd_num = 1000000
output = "/content/drive/MyDrive/Colab Notebooks/OS/test_1M_GET.input"
with open( output, 'w') as fout:
    for i in range(cmd_num):
        if ((i%10000) == 0):
            print(i)
        ctype = randint(1,6)
        cmd: str 
        # PUT
        if ctype < 7:
            cmd = f"PUT {randKey()} {randValue()}"
        # GET
        elif ctype < 10:
            cmd = f"GET {randKey()}"
        # SCAN
        else:
            key1, key2 = rand2Key(100) 
            cmd = f"SCAN {key1} {key2}"
        fout.write(cmd+"\n")
```
