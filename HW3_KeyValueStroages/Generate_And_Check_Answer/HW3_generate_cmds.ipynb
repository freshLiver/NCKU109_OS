```
from random import randint, choice

ALLOW_CHARS = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"

# key   : [0:2**63-1]
# value : always 128 digits of [A-Za-z0-9]

# PUT   : PUT key value
# GET   : GET key
# SCAN  : SCAN key1 key2
```

```
def randKey( init = 0 ) -> int:
    return randint( init, 2**63-1)

def rand2Key(limit = None) -> (int,int):
    key1 = randKey()
    if limit is None :
        key2 = randKey(key1+1)
    else:
        key2 = key1+limit
    return (key1,key2)

def randValue() -> str:
    cmd = ""
    for i in range(128):
        cmd += choice(ALLOW_CHARS)
    return cmd
```

```
cmd_num = 100
with open( f"sample{cmd_num}.input", 'w') as fout:
    for i in range(cmd_num):
        ctype = randint(0,2)
        cmd: str 
        # PUT
        if ctype == 0:
            cmd = f"PUT {randKey()} {randValue()}"
        # GET
        elif ctype == 1:
            cmd = f"GET {randKey()}"
        # SCAN
        else:
            key1, key2 = rand2Key(100) 
            cmd = f"SCAN {key1} {key2}"
        fout.write(cmd+"\n")
```
